```python
import random
import pandas as pd
import numpy as np
import copy
RANDOM_SEED = 42

class Synapse:
  def __init__(self, Wa=None, Wb=None):
    """Initialize a synapse with weights."""
    self.Wa = Wa  # Weight for input
    self.Wb = Wb  # Bias term or other weight (not used in synapse_output)

  def synapse_output(self, x):
    """Compute the output of the synapse given an input."""
    return self.Wa * x  # Weighted input

  def W_update(self, Wa=None):
    """Update the synapse weight."""
    self.Wb = copy.deepcopy(self.Wa)  # Update bias weight
    self.Wa = Wa  # Update weight

class Neuron:
  def __init__(self, limiar=None, n=None, synapses=[]):
    """Initialize the neuron with threshold, learning rate, and synapses."""
    self.limiar = limiar
    self.x = []  # x is a list of inputs, like [1, 2]
    self.y = []  # y is the expected output
    self.synapses = synapses  # list of synapses, e.g., [Synapse(), ...]
    self.n = n  # learning rate

  def start_synapses(self, number=None):
    """Initialize synapses with random weights."""
    random.seed(RANDOM_SEED)
    return [Synapse(random.uniform(0.0, 1.0), random.uniform(0.0, 1.0)) for _ in range(number)]

  def neurons_output(self, x=None):
    """Compute the output of the neuron."""
    if x:
      self.x = x
    # Sum weighted inputs from synapses
    value = sum([s.synapse_output(x_val) for x_val, s in zip(self.x, self.synapses)])
    value -= self.limiar  # Apply the threshold (bias)
    return 1 if value >= 0 else -1

  def sWeight_update(self, output=None):
    """Calculate the weight update using the Perceptron rule."""
    wb = [s.Wb for s in self.synapses]  # Bias weights
    wb.insert(0, self.limiar)  # Include bias in weight update
    wb = np.array(wb)

    x = [-1] + self.x  # Insert bias input (-1)
    x = np.array(x)

    # Calculate weight update
    weight_update = wb.T + self.n * (self.y - output) * x.T
    return weight_update

  def activate(self, x=[], y=[], synapses=[]):
    """Train the neuron with the given inputs and expected outputs."""
    self.synapses = self.start_synapses(len(x[0]))
    predictions = []
    for k,(x, y) in enumerate(zip(x,y)):
      self.x = x
      self.y = y
      output = self.neurons_output()
      predictions.append(output)
      print(k," - Neuron Output:", output, " - ", output==self.y)
      weight_update = self.sWeight_update(output)
      print("\033[93m",k," - Weight Update:\033[0m")
      print(weight_update)
      self.limiar = weight_update[0]
      weight_update = weight_update[1:]
      for w, s in zip(weight_update, self.synapses):
        s.W_update(w)
    return pd.Series(predictions)
```

```python
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd

iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

iris_df['species'] = iris.target

species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

iris_df.loc[iris_df['species'] == 'setosa', 'species'] = 1
iris_df.loc[iris_df['species'] != 1, 'species'] = -1

features = iris_df[iris.feature_names].values.tolist()
species = iris_df['species'].tolist()

neuron = Neuron(0.5,0.5)  # Initialize neuron with inputs, limiar, and learning rate
predictions = neuron.activate(features, species)  # Initialize synapses based on the number of inputs
print(predictions.value_counts())
print(classification_report(predictions, species))
labels = ['n-setosa','setosa']
cm = confusion_matrix(species, predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot()
plt.show()
```
